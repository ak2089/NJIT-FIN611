```python
# Machine Learning with Python using Linear Regression and Decison Tree
# Import necessary libraries
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
#data path 
df = pd.read_csv('data_ml.csv', parse_dates=['date'])
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
df.sort_values(by='date')
```

```python
#splitting data into two parts for Test and Train to test our model 

split_date = pd.to_datetime('01/01/2015')
df_train = df.loc[df['date'] <= split_date]
df_test = df.loc[df['date'] > split_date]
df_train
```

```python
df_test
```

```python
df_train=df_train.drop(columns=["R3M_Usd","R6M_Usd","R12M_Usd"])
df_test=df_test.drop(columns=["R3M_Usd","R6M_Usd","R12M_Usd"])
```

```python
predictors1 = df_train.drop(columns=["date"])
predictors2 = df_test.drop(columns=["date"])
```

```python
y_train = predictors1["R1M_Usd"]
y_test = predictors2["R1M_Usd"]
predictors_train= predictors1.drop(columns=["R1M_Usd"])
predictors_test= predictors2.drop(columns=["R1M_Usd"])
```

```python
import sklearn.preprocessing as sk
scaler = sk.StandardScaler()
predictors_train = scaler.fit_transform(predictors_train)
predictors_test = scaler.fit_transform(predictors_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(predictors_train,y_train)
```

```python
y_prediction = lr.predict(predictors_test)
```

```python
y_prediction2 = lr.predict(predictors_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#this is the test MSE
mean_squared_error(y_test, y_prediction)
```

```python
#this is the train MSE
mean_squared_error(y_train, y_prediction2)
```

```python
#this is the test r2
r2_score(y_test, y_prediction)
```

```python
#this is the train r2
r2_score(y_train, y_prediction2)
```
